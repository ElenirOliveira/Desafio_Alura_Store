# 3. Média de Avaliação das Lojas

```python

soma_avaliacoes = loja['Avaliação da compra'].sum()
media_avaliacoes = loja['Avaliação da compra'].mean()
print(f"Soma das avaliações das compras: {soma_avaliacoes}, Média: {media_avaliacoes:.2f}")

soma_avaliacoes = loja2['Avaliação da compra'].sum()
media_avaliacoes = loja2['Avaliação da compra'].mean()
print(f"Soma das avaliações das compras: {soma_avaliacoes}, Média: {media_avaliacoes:.2f}")

soma_avaliacoes = loja3['Avaliação da compra'].sum()
media_avaliacoes = loja3['Avaliação da compra'].mean()
print(f"Soma das avaliações das compras: {soma_avaliacoes}, Média: {media_avaliacoes:.2f}")

soma_avaliacoes = loja4['Avaliação da compra'].sum()
media_avaliacoes = loja4['Avaliação da compra'].mean()
print(f"Soma das avaliações das compras: {soma_avaliacoes}, Média: {media_avaliacoes:.2f}")
```

```python
# Calcula a soma e a média das avaliações para cada loja
medias = {}
for i, df in enumerate([loja, loja2, loja3, loja4], start=1):
	if 'Avaliação da compra' in df.columns:
		media = df['Avaliação da compra'].mean()
		medias[f'Loja {i}'] = media
	else:
		print(f"⚠️ Coluna 'Avaliação da compra' não encontrada em Loja {i}")

# Verifica se alguma média foi calculada
if not medias:
	print("❌ Nenhuma média foi calculada. Verifique se os DataFrames e a coluna 'Avaliação da compra' existem.")
else:
	# Cria a Series para as médias
	medias_series = pd.Series(medias).sort_values(ascending=False)

	# Cria um gráfico de barras
	plt.figure(figsize=(12, 6))  # Ajusta o tamanho do gráfico (largura x altura)
	medias_series.plot(kind='bar', color='cornflowerblue')

	plt.title('Média de Avaliação por Loja')
	plt.xlabel('Loja')
	plt.ylabel('Média das Avaliações')
	plt.ylim(0, 5)
	plt.xticks(rotation=0)
	plt.grid(axis='y', linestyle='--', alpha=0.7)
	plt.tight_layout()
	plt.show()

# Gráfico de pizza com a média de avaliações por loja
plt.figure(figsize=(8, 8))
cores = ['blue', 'pink', 'yellow', 'violet']

medias_series.plot(kind='pie',
                   autopct='%1.2f',
                   startangle=90,
                   colors=cores,
                   shadow=True,
                   wedgeprops={'edgecolor': 'black'},
                   labels=None)  # remove os rótulos no gráfico

plt.title('Média de Avaliação por Loja - Gráfico de Pizza')
plt.legend(labels=medias_series.index,
           title="Lojas",
           loc='center left',
           bbox_to_anchor=(1, 0.5))
plt.ylabel('')
plt.tight_layout()
plt.show()



```

# 5. Frete Médio por Loja


```python

# Calcula soma e média de frete por loja
soma1 = loja['Frete'].sum()
media1 = loja['Frete'].mean()

soma2 = loja2['Frete'].sum()
media2 = loja2['Frete'].mean()

soma3 = loja3['Frete'].sum()
media3 = loja3['Frete'].mean()

soma4 = loja4['Frete'].sum()
media4 = loja4['Frete'].mean()

# Mostra os valores no terminal
print(f"Loja 1 - Soma dos fretes: {soma1:.2f}, Média: {media1:.2f}")
print(f"Loja 2 - Soma dos fretes: {soma2:.2f}, Média: {media2:.2f}")
print(f"Loja 3 - Soma dos fretes: {soma3:.2f}, Média: {media3:.2f}")
print(f"Loja 4 - Soma dos fretes: {soma4:.2f}, Média: {media4:.2f}")

# Cria uma Series com as médias
medias_frete = pd.Series({
    'Loja 1': media1,
    'Loja 2': media2,
    'Loja 3': media3,
    'Loja 4': media4
}).sort_values(ascending=False)

# Cores e gráfico
cores = ['cornflowerblue', 'lightgreen', 'salmon', 'violet']

plt.figure(figsize=(10, 6))
bars = medias_frete.plot(kind='bar', color=cores, edgecolor='black')

# Adiciona valores nas barras
for i, valor in enumerate(medias_frete):
    plt.text(i, valor + 0.1, f'{valor:.2f}', ha='center', fontsize=10)

# Títulos e legendas
plt.title('Média do Frete por Loja')
plt.xlabel('Loja')
plt.ylabel('Média do Frete (R$)')
plt.ylim(0, max(medias_frete) + 5)
plt.xticks(rotation=0)

# Legenda personalizada
from matplotlib.patches import Patch
legenda = [Patch(color=cores[i], label=nome) for i, nome in enumerate(medias_frete.index)]
plt.legend(handles=legenda, title='Lojas', loc='upper right')

plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Criar um dicionário com IDs únicos para cada produto
produtos_unicos = {Produto: idx for idx, Produto in enumerate(loja2['Produto'].unique(), 1)}

# Mapear para a nova coluna
loja2['ID_Produto'] = loja2['Produto'].map(produtos_unicos)


# Criar uma nova coluna combinando ID_Produto e Produto
loja2['ID_Produto_Produto'] = loja2['ID_Produto'].astype(str) + " - " + loja2['Produto']

# Criar uma tabela dinâmica com Produto nas linhas, Data nas colunas e somando Avaliação da compra
# Substituímos 'Quantidade Vendida' por 'Avaliação da compra' porque 'Quantidade Vendida' não existe
# Agrupar por Produto e Data e adicionar ID
tabela_dinamica = loja2.pivot_table(
    index=['ID_Produto', 'Produto'],
    columns='Data da Compra',
    values='Avaliação da compra',
    aggfunc='count',
    fill_value=0
)
# Adicionar uma coluna com o total de vendas por produto
tabela_dinamica['Total de Vendas'] = tabela_dinamica.sum(axis=1)


# Mostrar a tabela final
print("Tabela dinâmica de vendas por produto, ID_Produto e data:")
print(tabela_dinamica)
print(loja2[['ID_Produto', 'Produto']].head())



```

```python
# Criar uma nova tabela só com ID, Produto e total de vendas
resumo_vendas = tabela_dinamica['Total de Vendas'].reset_index()
resumo_vendas = resumo_vendas.rename(columns={'Total de Vendas': 'Quantidade Vendida'})
total_vendas = resumo_vendas['Quantidade Vendida'].sum()
resumo_vendas['Total de Vendas'] = total_vendas


print("Resumo de vendas por produto com ID:")
print(resumo_vendas)
```
